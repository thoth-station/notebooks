# Discovering installed packages in Thoth

As a developer, you often want to know whick packages you have present in the environment as well as what are their versions so you know which features you can use or it can help you with tracing bugs. Thoth offers you an ability to automatically discover Python packages installed in the environment such as this Jupyter notebook is.

To discover packages that are present in the environment, you can simply import `packages_info()` function that is available in the `thoth-lab` package and run it:

```python
from thoth.lab import packages_info


packages_info()
```

The `packages_info()` function gives you a listing of all the available packages and their versions available for the Thoth project. It also states whether the given package can be imported without any errors.

The listing above gives you a list of Thoth's packages. This is the default behaviour of `packages_info()` function. However, if you would like to get listing of all the available packages installed in the current environment, you can do so by requesting it:

```python
df = packages_info(thoth_packages=False)
df
```

As you can see above, the optional parameter `thoth_packages` set to `False` forces discovery of all the available packages. As the return statement of `packages_info()` function is a Pandas [DataFrame](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.html), you can also perform queries on results. To check what version of Pandas is available, just query the resulting DataFrame:

```python
df.loc[df['package'] == 'pandas']
```
